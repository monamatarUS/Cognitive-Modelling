```python
# Setup the environment
import nengo
import nengo.spa as spa

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
D = 64
model = spa.SPA()

with model:
    model.visual = spa.State(D)
    model.speach = spa.State(D)
    model.hand = spa.State(D)
    model.phrase = spa.State(D)
    model.verb = spa.State(D, feedback=1)
    model.noun = spa.State(D, feedback=1)  
    
    vocab = model.get_output_vocab('visual')
    actions = spa.Actions(
        'dot(visual, SAY+WRITE) --> verb=visual',
        'dot(visual, HELLO+GOODBYE) --> noun=visual',
        'dot(phrase, SAY*VERB) - dot(visual, HELLO+GOODBYE+SAY+WRITE) --> speach=phrase*~NOUN',
        'dot(phrase, WRITE*VERB) - dot(visual, HELLO+GOODBYE+SAY+WRITE) --> hand=phrase*~NOUN',
    )
    model.bg = spa.BasalGanglia(actions)
    model.thal = spa.Thalamus(model.bg)
    cortical_actions = spa.Actions(
        'phrase = verb*VERB + noun*NOUN',
    )
model.cortical = spa.Cortical(cortical_actions)
```

```python
def input_vision(t):
    sequence = 'SAY HELLO NONE WRITE GOODBYE NONE GOODBYE SAY NONE'.split()
    index = int(t / 0.5) % len(sequence)
    return sequence[index]

with model:
     model.input = spa.Input(visual=input_vision)
```

```python
with model:
    visual = nengo.Probe(model.visual.output, synapse=0.03)
    phrase = nengo.Probe(model.phrase.output, synapse=0.03)
    hand = nengo.Probe(model.hand.output, synapse=0.03)
    speach = nengo.Probe(model.speach.output, synapse=0.03)
    noun = nengo.Probe(model.noun.output, synapse=0.03)
    verb = nengo.Probe(model.verb.output, synapse=0.03)
    actions = nengo.Probe(model.thal.actions.output, synapse=0.01)
    utility = nengo.Probe(model.bg.input, synapse=0.01)
```

```python
sim = nengo.Simulator(model)    # Create the simulator object
sim.run(4.5)    # Simulate the model for 4.5 seconds
```

```python
fig = plt.figure(figsize=(16,16))
p1 = fig.add_subplot(8,1,1)
p1.plot(sim.trange(), model.similarity(sim.data, visual))
#p1.legend(model.get_output_vocab('visual').keys, fontsize='x-small')
p1.legend(model.get_output_vocab('visual').keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=9)
p1.set_ylabel('Visual')

p2 = fig.add_subplot(8,1,2)
p2.plot(sim.trange(), model.similarity(sim.data, phrase))
#p2.legend(model.get_output_vocab('phrase').keys, fontsize='x-small')
p2.legend(model.get_output_vocab('phrase').keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=9)
p2.set_ylabel('Phrase')

p3 = fig.add_subplot(8,1,3)
p3.plot(sim.trange(), model.similarity(sim.data, hand))
#p3.legend(model.get_output_vocab('hand').keys, fontsize='x-small')
p3.legend(model.get_output_vocab('hand').keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=9)
p3.set_ylabel('Hand')

p4 = fig.add_subplot(8,1,4)
p4.plot(sim.trange(), model.similarity(sim.data, speach))
#p4.legend(model.get_output_vocab('speach').keys, fontsize='x-small')
p4.legend(model.get_output_vocab('speach').keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=9)
p4.set_ylabel('Speach')

p5 = fig.add_subplot(8,1,5)
p5.plot(sim.trange(), model.similarity(sim.data, noun))
#p5.legend(model.get_output_vocab('noun').keys, fontsize='x-small')
p5.legend(model.get_output_vocab('noun').keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=9)
p5.set_ylabel('Noun')

p6 = fig.add_subplot(8,1,6)
p6.plot(sim.trange(), model.similarity(sim.data, verb))
#p6.legend(model.get_output_vocab('verb').keys, fontsize='x-small')
p6.legend(model.get_output_vocab('verb').keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=9)
p6.set_ylabel('Verb')

p7 = fig.add_subplot(8,1,7)
p7.plot(sim.trange(), sim.data[actions])
p7.set_ylabel('Action')

p8 = fig.add_subplot(8,1,8)
p8.plot(sim.trange(), sim.data[utility])
p8.set_ylabel('Utility')

fig.subplots_adjust(hspace=0.2)
```
